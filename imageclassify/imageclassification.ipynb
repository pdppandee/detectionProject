# PRE PROCESSING

```python
data = data.map(lambda x,y: (x/255, y)) 
#.map transformation data(from disk) in pipeline
#x = image
#y = label
#scaling by /255
```

```python
data.as_numpy_iterator().next()
```

# Split data

```python
#Split data
train_size = int(len(data)*.7) #70 percent of data
val_size = int(len(data)*.2) +1 #validate the module while running 
test_size = int(len(data)*.1)+1
```

```python
print("length of data:", len(data))
print("train_size:", train_size)
print("validate size:", val_size)
print("test size:", test_size)
```

train size + val size + test size = length of data

```python
train = data.take(train_size)
val = data.skip(train_size).take(val_size) #skip the train batch and take val
test = data.skip(train_size+val_size).take(test_size)
```

```python
train
```

# DEEP LEARNING

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Dense, Flatten, Dropout
```

```python
model = Sequential()
```

```python
#MaxPooling2D??
```

```python
#input layer
#Con2D(16 filters),(size of filter),move ทีละ 1 pixel, 
model.add(Conv2D(16, (3,3), 1, activation='relu', input_shape=(256,256,3)))
model.add(MaxPooling2D())

model.add(Conv2D(32, (3,3), 1, activation='relu'))
model.add(MaxPooling2D())

model.add(Conv2D(16, (3,3), 1, activation='relu'))
model.add(MaxPooling2D())

model.add(Flatten())
#dense: fully connected layars -> 256 neuron
model.add(Dense(256, activation='relu'))
# model.add(Dense(256, activation='relu'))
model.add(Dense(1, activation='sigmoid')) #output ถูกจัดให้อยู่ในเรนจ์ 0-1
```

# TRAIN

```python
logdir='logs'
```

```python
tensorboard_callback = tf.keras.callbacks.TensorBoard(log_dir=logdir)
```

```python
hist = model.fit(train, epochs=20, validation_data=val, callbacks=[tensorboard_callback])
```

```python
hist.history
```

```python
fig = plt.figure()
plt.plot(hist.history['loss'], color='teal', label='loss')
plt.plot(hist.history['val_loss'], color='orange', label='val_loss')
fig.suptitle('Loss', fontsize=20)
plt.legend(loc="upper left")
plt.show()
```

```python
fig = plt.figure()
plt.plot(hist.history['accuracy'], color='teal', label='accuracy')
plt.plot(hist.history['val_accuracy'], color='orange', label='val_accuracy')
fig.suptitle('Accuracy', fontsize=20)
plt.legend(loc="upper left")
plt.show()
```

# Evaluate Performane

```python
from tensorflow.keras.metrics import Precision, Recall, BinaryAccuracy
```

```python
pre = Precision()
re = Recall()
acc = BinaryAccuracy()

print("length of test data:", len(test))
```

```python
for batch in test.as_numpy_iterator(): 
    X, y = batch
    yhat = model.predict(X) #will return value 0-1 drom sigmoid
    pre.update_state(y, yhat)
    re.update_state(y, yhat)
    acc.update_state(y, yhat)
```

```python
#high value = good / max height = 1 
print(f'Precision:{pre.result().numpy()}, Recalll:{re.result().numpy()}, Accuracy:{acc.result().numpy()}')
```
